```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException

browser = webdriver.Chrome()
browser.get('https://www.cathaybk.com.tw/cathaybk/')

delay = 3 # seconds
try:
    myElem = WebDriverWait(browser, delay).until(EC.presence_of_element_located((By.ID, 'searchBox')))
    print('網頁下載完成')
except TimeoutException:
    print("網頁下載時間過長")


# 設定瀏覽器視窗大小
width = 700 # in pixels
height = 1200 # in pixels
browser.set_window_size(width, height)

# 截圖
browser.get_screenshot_as_file('screen_1.png')

#漢堡選單
burgerMenu = browser.find_element(By.CLASS_NAME, "cubre-a-burger")
burgerMenu.click()

#level 1 menu - 產品介紹 
l1Menu = browser.find_element(By.XPATH, '/html/body/div[1]/header/div/div[3]/div/div[2]/div[1]/div/div[1]')
l1Menu.click()

#level 2 menu - 信用卡
l2Menu = browser.find_element(By.XPATH, '/html/body/div[1]/header/div/div[3]/div/div[2]/div[1]/div/div[1]/div[2]/div/div[1]')
l2Menu.click()

delay = 3 # seconds
try:
    myElem = WebDriverWait(browser, delay).until(EC.presence_of_element_located((By.CLASS_NAME, 'cubre-a-menuLink')))
    print('網頁下載完成')
except TimeoutException:
    print("網頁下載時間過長")

#計算有幾項目在信用卡選單下面
#section = 'body > div.cubre-o-container > header > div > div.cubre-o-header__nav > div > div.cubre-o-nav__content > div.cubre-o-nav__menu > div > div.cubre-o-menu__item.is-L1open > div.cubre-o-menu__content > div > div.cubre-o-menuLinkList__item.is-L2open'
selectionItems = browser.find_element(By.XPATH,'/html/body/div[1]/header/div/div[3]/div/div[2]/div[1]/div/div[1]/div[2]/div/div[1]').find_elements(By.CSS_SELECTOR, 'a#lnk_Link.cubre-a-menuLink') 
print('信用卡選單下面共有 %d 個選項。' % len(selectionItems))

import time
time.sleep(5)
browser.get_screenshot_as_file('menu_1.png')

#卡片介紹 
cardInfoMenu = browser.find_element(By.LINK_TEXT, "卡片介紹")
cardInfoMenu.click()
time.sleep(1)

#停發卡
stoppedCard = browser.find_element(By.XPATH, '/html/body/div[1]/main/article/div/div/div/div[1]/div/div/a[6]')
stoppedCard.click()
time.sleep(5)

#計算停發卡數量
cardTitle = browser.find_element(By.XPATH, '/html/body/div[1]/main/article/section[6]/div/div[2]').find_elements(By.CLASS_NAME, 'cubre-m-compareCard__title')
print('共發現停發卡片 %d 張' % len(cardTitle))

# -*- coding: utf-8 -*-
import os
from os.path import isfile, isdir, join

files = os.listdir('./card')
for fileIndex in range(len(files)):
#    print('file: %s' % files[fileIndex])
    rmfile = './card/' + files[fileIndex]
    os.remove(rmfile)

#抓第一張圖
browser.get_screenshot_as_file('./card/card_1.png')

#依序點開抓其他圖
for i in range(2,len(cardTitle)+1):
    dotXPath = '/html/body/div[1]/main/article/section[6]/div/div[2]/div/div[2]/span[' + str(i) + ']'
    nextDot = browser.find_element(By.XPATH, dotXPath)
    nextDot.click()
    time.sleep(5)
    fileName = './card/card_' + str(i) + '.png'
    browser.get_screenshot_as_file(fileName)

#驗證抓圖數量
# 以迴圈處理
initial_count = 0
dir = "./card"
for path in os.listdir(dir):
    if os.path.isfile(os.path.join(dir, path)):
        initial_count += 1
print(initial_count)

print('共有 %d 張螢幕截圖' % initial_count)

if initial_count == len(cardTitle):
    print('卡片截圖數量為 %d，與之前步驟清查的數量 %d 相符。' % (initial_count,len(cardTitle)))
else:
    print('卡片截圖數量為 %d，與之前步驟清查的數量 %d 不符。' % (initial_count,len(cardTitle)))
    
browser.close()
```
